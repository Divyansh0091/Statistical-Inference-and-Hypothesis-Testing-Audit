```python
import numpy as np 

import scipy.stats as stat

new_value = np.array([110,125,160,130,160,175,190,200,210,270])
com = 160
stat_value,p_value = stat.ttest_1samp(new_value,com) 
print(f"T-Statisti{stat_value}")
print(f"P-Value{p_value}")
alpha = 0.05 
if p_value > alpha:
    print("Result: Hypothesis Rejected")
else:
    print("Result: Hypothesis not Rejected")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,4))
sns.histplot(data=new_value,kde=True,color='blue')
plt.xlabel("Total Value",fontsize=10,color='red')
plt.ylabel("Comparition",fontsize=10,color='green')
plt.axvline(com,color='red',linestyle='--')
plt.title("New comparition",fontsize=10,color='red')
plt.show()
```
